```python
import pickle
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.DataFrame(pd.read_pickle('../models/baseline_100D_history'))
```

```python
file_list = ['baseline_50D_history', 'baseline_100D_history', 'baseline_200D_history', 'baseline_300D_history']

histories = {}
for file in file_list:
    
    with open('../models/' + file, 'rb') as f:
        training_dict = pickle.load(f)
    histories[file.split('_')[1]] = training_dict
```

```python
for key in histories.keys():
    
    training_dict = histories[key]
    n_epochs = len(training_dict['loss'])

    with plt.style.context('seaborn-talk'):

        ax = plt.axes()
        ax.plot(training_dict['loss'], label=key, marker='^')
        #ax.plot(training_dict['val_loss'], label='Val Loss', marker='s')
        ax.xaxis.set_major_locator(plt.MaxNLocator(n_epochs))
        ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, y: int(x + 1)))
        ax.tick_params(axis="x", labelsize=18)
        ax.tick_params(axis="y", labelsize=15)
        plt.title('Training Loss', fontsize=25, y=1.05)
        plt.xlabel('Epoch', fontsize=20)
        plt.ylabel('Loss', fontsize=20)
        plt.legend(fontsize=20)
```
